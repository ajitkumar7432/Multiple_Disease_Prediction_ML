Visualizations


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Count class labels in Parkinson's dataset
counts = parkinsons_data['status'].value_counts()
labels = ['Healthy (0)', 'Parkinson’s (1)']

# Plot pie chart
plt.figure(figsize=(5, 5))
plt.pie(counts, labels=labels, autopct='%1.1f%%', colors=['lightgreen', 'orange'], startangle=140)
plt.title('Parkinson’s Status Distribution')
plt.show()
```

```python
# For Parkinson's Dataset
important_features = ['spread1', 'PPE', 'RPDE', 'DFA']

for col in important_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=parkinsons_data, x=col, hue='status', kde=True, palette='husl')
    plt.title(f'{col} Distribution by Status (Parkinson\'s)')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(12, 10))  # Larger size for more features
sns.heatmap(parkinsons_data.drop(columns='name').corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Feature Correlation Heatmap - Parkinson’s Dataset")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.scatterplot(data=parkinsons_data, x='spread1', y='PPE', hue='status', palette='Set1', alpha=0.8)

plt.title('Best Feature Pair for Linear Separability (spread1 vs PPE)')
plt.xlabel('spread1')
plt.ylabel('PPE')
plt.grid(True)
plt.show()
```

Separating the features & Target

```python
X = parkinsons_data.drop(columns=['name','status'], axis=1)
Y = parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data & Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

Model Training

Support Vector Machine Model

```python
model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy score on training data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of test data : ', test_data_accuracy)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Predict on test set
Y_pred_parkinson = model.predict(X_test)

# Generate confusion matrix
cm_parkinson = confusion_matrix(Y_test, Y_pred_parkinson)

# Plot confusion matrix
disp_parkinson = ConfusionMatrixDisplay(confusion_matrix=cm_parkinson, display_labels=model.classes_)
disp_parkinson.plot(cmap='Purples')
plt.title("Confusion Matrix - Parkinson’s Disease Prediction")
plt.grid(False)
plt.show()
```

Building a Predictive System

```python
input_data = (197.07600,206.89600,192.05500,0.00289,0.00001,0.00166,0.00168,0.00498,0.01098,0.09700,0.00563,0.00680,0.00802,0.01689,0.00339,26.77500,0.422229,0.741367,-7.348300,0.177551,1.743867,0.085569)

# changing input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the data
std_data = scaler.transform(input_data_reshaped)

prediction = model.predict(std_data)
print(prediction)


if (prediction[0] == 0):
  print("The Person does not have Parkinsons Disease")

else:
  print("The Person has Parkinsons")
```
