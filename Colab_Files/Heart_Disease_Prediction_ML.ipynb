Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
# loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('/content/heart.csv')
```

```python
# getting some info about the data
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
# statistical measures about the data
heart_data.describe()
```

```python
# checking the distribution of Target Variable
heart_data['target'].value_counts()
```

0 --> Healthy Heart

Visualisations

```python
import matplotlib.pyplot as plt

# Count target classes
counts = heart_data['target'].value_counts()
labels = ['No Heart Disease (0)', 'Heart Disease (1)']

# Plot pie chart
plt.figure(figsize=(5, 5))
plt.pie(counts, labels=labels, autopct='%1.1f%%', colors=['skyblue', 'salmon'], startangle=140)
plt.title('Target Class Distribution - Heart Disease')
plt.show()
```

```python
import seaborn as sns
# For Heart Disease Dataset
important_features = ['thalach', 'oldpeak', 'cp', 'age']

for col in important_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=heart_data, x=col, hue='target', kde=True, palette='husl')
    plt.title(f'{col} Distribution by Target (Heart Disease)')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(heart_data.corr(), annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Feature Correlation Heatmap - Heart Disease Dataset")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot of two features from the heart dataset
plt.figure(figsize=(8, 6))
sns.scatterplot(data=heart_data, x='age', y='thalach', hue='target', palette='Set1', alpha=0.8)

plt.title('Visual Check for Linear Separability (Age vs Thalach)')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.grid(True)
plt.show()
```

```python
selected_features = ['age', 'thalach', 'oldpeak', 'cp']

# Pairplot with hue based on target
sns.pairplot(heart_data, hue='target', vars=selected_features, palette='Set1', plot_kws={'alpha': 0.7})
plt.suptitle('Feature Pairwise Plots for Separability Check (Heart Disease)', y=1.02)
plt.show()
```

Splitting the Features and Target

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Predict on test set
Y_pred_heart = model.predict(X_test)

# Generate confusion matrix
cm_heart = confusion_matrix(Y_test, Y_pred_heart)

# Plot confusion matrix
disp_heart = ConfusionMatrixDisplay(confusion_matrix=cm_heart, display_labels=model.classes_)
disp_heart.plot(cmap='Reds')
plt.title("Confusion Matrix - Heart Disease Prediction")
plt.grid(False)
plt.show()
```

Building a Predictive System

```python
input_data = (62,0,0,140,268,0,0,160,0,3.6,0,2,2)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```
